**Stemming**

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
sentences = nltk.sent_tokenize(paragraph)
stemmer = PorterStemmer()
```

```python
# checking stop words in english 
stopwords.words('english')
```

```python
len(stopwords.words('english'))
```

```python
len(sentences)
```

```python
# stemming 
for i in range (len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words)
```

```python
# cleaning the text
import re # regular expression library
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
ps = PorterStemmer()
wordnet = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentences)):
  review = re.sub('[^a-zA-Z]', ' ', sentences[i])
  review = review.lower()
  review = review.split()
  #review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))] # stemming
  review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))] # lemmatizing
  review = ' '.join(review)
  corpus.append(review)
```

```python
# creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
```
